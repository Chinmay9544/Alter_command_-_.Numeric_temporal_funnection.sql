**1. What is the average median income of the data set and check the distribution of data using appropriate plots. Please explain the distribution of the plot.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset into a pandas DataFrame
data = pd.read_excel('carhouse.xlsx')

# Calculate the average median income
average_median_income = np.mean(data['median_income'])

# Plot the distribution of data using a histogram
plt.hist(data['median_income'], bins=20)
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.title('Distribution of median_income')
plt.show()

# Plot the distribution of data using a box plot
plt.boxplot(data['median_income'])
plt.ylabel('Income')
plt.title('Distribution of median_income')
plt.show()

# Print the average median income
print("Average median_income:", average_median_income)
```

**EXPLANATION**

The first plot is a histogram that shows the frequency of income values within different bins. The second plot is a box plot that displays the quartiles, median, and potential outliers in the data.

The plot distribution of this median income lies between 3-5

Most of the data lies between 25%-75% of the distribution

By examining these plots, you can understand the distribution of the median income data. The histogram can indicate whether the data is symmetric or skewed, while the box plot provides information on the spread and presence of outliers.

**10.Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.**

```python

# Define a function to categorize total_bedrooms size
def categorize_bedroom_size(total_bedrooms):
    if total_bedrooms <= 10:
        return 'small'
    elif 11 <= total_bedrooms < 1000:
        return 'medium'
    else:
        return 'large'

# Apply the function to create the 'total_bedroom_size' column
data['total_bedroom_size'] = data['total_bedrooms'].apply(categorize_bedroom_size)

# Display the modified dataset
print(data)
```
